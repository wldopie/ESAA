### **2) DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환**



##### a. 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기
  - DataFrame으로 변환 시 칼럼명 지정
  - DataFrame은 행과 열을 갖는 2차원 데이터이므로, 2차원 이하의 데이터들만 DataFrame으로 변환 가능

```python
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape, '\n')

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1, '\n')

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
# 3개의 칼럼명이 필요함
col_name2 = ['col1', 'col2', 'col3']

# 2x3 형태의 리스트와 ndarray 생성한 뒤, 이를 DataFrame으로 변환
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape, '\n')

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2, '\n')

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
# 딕셔너리를 DataFrame으로 변환 (dic = {key : value})
# key는 문자열 칼럼명으로 매핑, value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}      # {칼럼명 : 칼럼 데이터}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

##### b. DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
  - **`넘파이 ndarray`** : DataFrame 객체의 **`values`**를 이용해 가능
  - **`리스트`**로의 변환 : values로 얻은 ndarray에 **`tolist()`** 호출
  - **`딕셔너리`**로의 변환 : DataFrame 객체의 **`to_dict()`** 호출
    - 인자로 'list'를 입력하면 딕셔너리의 값(value)이 리스트형으로 반환 

```python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
```

```python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### **4) DataFrame의 데이터 삭제**
  - **`DataFrame.drop`**(labels=None, axis=0, inplace=False)
    - i) **`axis`**
      - axis=0 : 로우 삭제, axis=1 : 칼럼 삭제
    - ii) **`inplace`**
      - **`inplace=False`** : 자기 자신의 DataFrame의 데이터는 삭제하지 않으며, **`삭제된 결과 DataFrame을 반환`** (**`원본 DataFrame은 그대로 유지`**)
      - **`inplace=True`** : 자기 자신의 DataFrame의 데이터를 삭제 (원본 DataFrame의 변화)
        - `True`로 설정할 경우 `반환값이 None`

```python
# Age_0, Age_by_10, Family_No 칼럼 삭제
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print('inplace=True로 drop 후 반환된 값:', drop_result)     # 
titanic_df.head(3)
```

### **7) 정렬, Aggregation 함수, GroupBy 적용**

##### a. DataFrame, Series의 **`정렬`** : **`sort_values()`**
  - 주요 입력 파라미터
    - **`by`** : 특정 칼럼을 입력하면 **`해당 칼럼으로 정렬`** 수행
    - **`ascending = True`** : **`오름차순`** 정렬 / **`ascending = False`** : **`내림차순`** 정렬
    - **`inplace = True`** : **`원본 데이터프레임`**을 정렬 / **`inplace = False`** : **`원본`** 데이터프레임은 **`그대로`** 유지, **`정렬된 결과값`**만을 반환

```python
# titanic_df의 Name 칼럼을 오름차순 정렬
titanic_sorted = titanic_df.sort_values( by=['Name'] )      # default : ascending = True
titanic_sorted.head(3)
```

```python
# Pclass와 Name 칼럼을 내림차순으로 정렬
titanic_sorted = titanic_df.sort_values( by=['Pclass', 'Name'], ascending=False)
titanic_sorted.head(3)
```

##### b. **`Aggregation`** 함수 적용
  - **`min()`**, **`max()`**, **`sum()`**, **`count()`**
  - DataFrame에 바로 aggregation 함수를 호출할 경우, 모든 칼럼에 해당 aggregation을 적용

```python
titanic_df.count()      # 모든 칼럼의 개수를 출력
```

```python
titanic_df[ ['Age', 'Fare'] ].mean()      # Age, Fare 칼럼에만 aggregation 함수를 적용
```

##### c. **`groupby()`** 적용
  - **`groupby()`** : 입력 파라미터 **`by에 칼럼을 입력`**하면 **`대상 칼럼으로 groupby`** (by는 생략 가능)
  - DataFrame.groupby().**`agg()`** : **`서로 다른 aggregation 함수를 적용`**시킬 경우 사용

```python
# Pclass 칼럼을 기준으로 Groupby된 새로운 DataFrame Groupby 객체를 반환
titanic_groupby = titanic_df.groupby(by='Pclass')
print(type(titanic_groupby))      # 기존의 데이터프레임에서 Groupby된 새로운 'DataFrame Groupby' 객체 반환
```

### **9) `apply lambda` 식으로 데이터 가공**
  - **`lambda`**  : 함수의 선언과 함수 내의 처리를 **`한 줄의 식으로 쉽게 변환`**하는 식
  - 구조
    - lambda (**`입력 인자`**) : (입력 인자를 기반으로 한 **`계산식`**)
  - **`map()`** : **`여러 개의 값을 입력 인자`**로 사용할 경우

```python
# 기존 방식의 함수 선언과 함수 내 처리
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
```

```python
# lambda 식을 활용한 계산
lambda_square = lambda x : x**2
print('3의 제곱은:', lambda_square(3))
```

```python
# 여러 개의 값을 입력 인자로 사용
a = [1,2,3]
squares = map(lambda x : x**2, a)     # map(lambda 식, 여러 개의 입력 인자)
list(squares)                         # map 함수를 적용할 경우, list() 형태로 출력
```

```python
# 문자열의 길이를 반환하는 칼럼 : Name_len
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[ ['Name', 'Name_len'] ].head(3)
```

```python
# 나이가 15세 미만이면 Child, 그렇지 않으면 Adult로 구분하는 새로운 칼럼 : Child_Adult
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x<15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(3)
```

```python
# 나이가 15세 이하이면 Child, 15~60세 사이는 Adult, 61세 이상은 Elderly로 구분하는 새로운 칼럼 : Age_Cat
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_Cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat = ''
  if age <=5:
    cat = 'Baby'
  elif age<=12:
    cat = 'Child'
  elif age<=18:
    cat = 'Teenager'
  elif age<=25:
    cat = 'Student'
  elif age<=35:
    cat = 'Young Adult'
  elif age<=60:
    cat = 'Adult'
  else:
    cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_Cat']].head()
```
