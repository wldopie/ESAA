### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv).

### Step 3. Assign it to a variable called df.

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv', sep=',')
print(df.head())
print(df.shape)
```

### Step 4. For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column

```python
# loc[] : 명칭 인덱싱
df = df.loc[:, 'school':'guardian']
df
```

### Step 5. Create a lambda function that will capitalize strings.

```python
lambda_capital = lambda x : x.upper()       # string.upper() : string을 대문자로 반환하는 함수
lambda_capital('abc')
```

### Step 6. Capitalize both Mjob and Fjob

```python
# Mjob Capitalizing
df['Mjob'].apply(lambda x : lambda_capital(x))

# Fjob Capitalizing
df['Fjob'].apply(lambda x : lambda_capital(x))
```

### Step 7. Print the last elements of the data set.

```python
df.tail(1)
```

### Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

```python
# original dataframe의 Mjob, Fjob이 lowercase인 이유는, step 6에서 대문자로 변환된 열을 original dataframe의 Mjob, Fjob으로 지정해주지 않고 단순히 결과만 출력했기 때문이다. 
df['Mjob'] = df['Mjob'].apply(lambda x : lambda_capital(x))
df['Fjob'] = df['Fjob'].apply(lambda x : lambda_capital(x))
df.tail(1)      # Mjob, Fjob 모두 대문자로 변환된 것을 확인할 수 있음
```

### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
# 17살보다 많으면 True, 그렇지 않으면 False를 반환하는 함수
# lambda 함수를 활용해 majority 정의

majority = lambda x : True if x > 17 else False
df['legal_drinker'] = df['age'].apply(lambda x : majority(x))
df.head()     # legal_drinker라는 열이 새롭게 형성되었고, 17살보다 많으면 True, 그렇지 않으면 False로 올바르게 반환
```
