### 4-1. 데이터 시각화가 필요한 이유

> a. **`앤스콤 4분할`** 그래프 살펴보기
  
  - 앤스콤이 지적한 함정과 데이터 시각화의 필요성
    - 데이터 그룹의 `수칫값이나 상관관계, 회귀선이 같다`는 이유로 `각 그룹의 데이터가 모두 같을 것이라고 착각`하는 경우

```python
# 5. 점 그래프 그리기 : plot(x, y, 'o')
plt.plot(dataset_1['x'], dataset_1['y'], 'o')
```

### 4-2. ***`matplotlib 라이브러리`*** 자유자재로 사용하기

> a. `기초 그래프` 그리기 - **`히스토그램`**, **`산점도 그래프`**, **`박스 그래프`**

```python
# 1. tips 데이터 집합을 불러와 변수 tips에 저장하기
tips = sns.load_dataset('tips')
print(tips.head())      # 지불 금액, 팁, 성별, 흡연 유무, 요일, 시간, 전체 인원
print(type(tips))
```

```python
# 2. 그래프의 기본 틀(fig)을 마련하고 그래프 격자(axes1) 구성하기
fig = plt.figure()

axes1 = fig.add_subplot(1,1,1)
```

```python
# 3. total_bill 칼럼으로 히스토그램 그리기 : hist()
# 변수를 하나만 사용하므로, "일변량 그래프" 또는 "단변량 그래프"
axes1.hist(tips['total_bill'], bins=10)       # x축의 간격이 10인 히스토그램
axes1.set_title('Histogram of Total Bill')    # 그래프 제목 설정
axes1.set_xlabel('Frequency')                 # x축 제목 설정
axes1.set_ylabel('Total Bill')                # y축 제목 설정

fig
```

```python
# 4. total_bill, tip 칼럼으로 산점도 그래프 그리기 : scatter()
# 두 가지의 변수를 사용하므로, "이변량 그래프"

scatter_plot = plt.figure()                             # 그래프 기본 틀 설정   
axes1 = scatter_plot.add_subplot(1,1,1)           

axes1.scatter(tips['total_bill'], tips['tip'])         # x = total_bill, y = tips인 산점도 
axes1.set_title('Scatterplot of Total Bill vs. Tip')    # 그래프 제목 설정
axes1.set_xlabel('Total Bill')                          # x축 제목 설정
axes1.set_ylabel('Tip')                                 # y축 제목 설정

# 왜 산점도 그래프는 위에서와 같이 "fig"를 입력하지 않아도 출력되는지?
```

```python
# 5. 이산형 변수(sex)와 연속형 변수(tip)를 이용하여 박스 그래프 그리기 : boxplot()

boxplot = plt.figure()                # 그래프의 기본 틀 설정
axes1 = boxplot.add_subplot(1,1,1)   

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],     # 여성인 사람들의 tip 값 추출
               tips[tips['sex'] == 'Male']['tip']],      # 남성인 사람들의 tip 값 추출
               labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

> b. **`다변량 그래프`** 그리기
  
  - `다변량 그래프`
    - **`3개 이상의 변수`**를 사용한 그래프
  - 다변량 그래프가 필요한 이유?
    - 적재적소에 맞는 그래프 요소를 추가하여 3개 이상의 변수를 잘 표현할 수 있도록 하는 그래프를 출력하기 위함

```python
# 1. 앞에서 실습한 산점도 그래프에 성별을 새 변수로 추가하기
# 성별은 서로 다른 색으로 구분
# Male, Female을 각각 1, 0으로 변환하는 함수
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
# 2. recode_sex가 반환한 값을 데이터프레임에 새로운 열로 추가
tips['sex_color'] = tips['sex'].apply(recode_sex)       # apply 함수 사용
```

```python
# 3. 테이블당 인원 수를 산점도 그래프에 추가하기 
# 테이블당 인원 수는 점의 크기로 구분

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 10,
    c = tips['sex_color'],
    alpha = 0.5             # 점의 투명도를 조절하는 파라미터
)

axes1.set_title('Total Bill vs. Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### 4-3. ***`seaborn 라이브러리`*** 자유자재로 사용하기

> a. **`단변량 그래프`** 그리기 - **`히스토그램`**

```python
# 1. tips 데이터 집합 불러오기
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
# 2. seaborn 라이브러리로 히스토그램 그리기 : subplots(), distplots()
# subplot() : 기본 틀 형성 / distplot() : 히스토그램, 밀집도 그래프 출력

ax = plt.subplots()                                      
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# 3. distplot() 메서드에서 히스토그램만 출력하기

ax = plt.subplots()                                      
ax = sns.distplot(tips['total_bill'], kde=False)          # kde = False : 밀집도 그래프를 제외하고 출력
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 4. distplot() 메서드에서 밀집도 그래프만 출력

ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'], hist=False)         # hist = False : 밀집도 그래프만 출력
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
# 5. distplot() 메서드에 rug 인자 추가하기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
# 6. count 그래프 출력하기 : countplot()
ax = plt.subplots( )
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency')
```

> b. 다양한 종류의 **`이변량 그래프`** 그리기

```python
# 1. 산점도 그래프 그리기 : regplot()을 통해 산점도 그래프, 회귀선 출력
ax = plt.subplots( )
ax = sns.regplot(x = 'total_bill', y = 'tip', data=tips)    # 산점도, 회귀선 모두 출력 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
# 산점도 그래프만 출력하려면 fit_reg = False로 지정
ax = plt.subplots( )
ax = sns.regplot(x = 'total_bill', y = 'tip', data=tips, fit_reg=False)   # 산점도만 출력
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
# 2. 산점도 그래프와 히스토그램을 한 번에 그리기 : jointplot(x, y, data)
joint = sns.jointplot(x = 'total_bill', y = 'tip', data=tips) 
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip') 
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 3. 육각형 그래프 그리기 : jointplot(x,y,data, kind='hex')
# 산점도 그래프의 경우 점이 겹치면 구분하기 어려우므로, 이를 보완하는 그래프
joint = sns.jointplot(x = 'total_bill', y = 'tip', data=tips, kind='hex') 
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip') 
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 4. 이차원 밀집도 그리기 : kdeplot()
ax = plt.subplots( )
ax = sns.kdeplot(data = tips['total_bill'],
                 data2 = tips['tip'],
                 shade=True)            # 그래프에 음영 효과를 줄 수 있음

ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
# 5. 바(bar) 그래프 그리기 : barplot()
# 바 그래프는 지정한 변수의 평균을 계산하여 그릴 수 있음

ax = plt.subplots( )
ax = sns.barplot(x = 'time', y = 'total_bill', data=tips) 
ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

```python
# 6. 박스 그래프 그리기 : boxplot()
# 박스 그래프 : min, Q1, median, Q3, max, outlier 등 다양한 통계량을 한 번에 표현
ax = plt.subplots( )
ax = sns.boxplot(x = 'time', y = 'total_bill', data=tips) 
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
# 7. 바이올린 그래프 그리기 : violinplot()
# 박스 그래프는 데이터의 분산이 모호하게 표현되므로, 이를 명확하게 하기 위한 그래프

ax = plt.subplots( )
ax = sns.violinplot(x = 'time', y = 'total_bill', data=tips) 
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
# 8. 관계 그래프 그리기 : pairplot(dataframe)
fig = sns.pairplot(tips)
```

```python
# 9. 관계 그래프의 중복성을 제거하고 원하는 그래프로 채워넣기 : PairGrid(dataframe)
# map_upper : 대각선을 중심으로 위쪽에 그릴 그래프 지정 (산점도 그래프)
# map_lower : 대각선을 중심으로 아래쪽에 그릴 그래프 지정 (이차원 밀집도)
# map_diag : 대각선을 중심으로 그래프 출력 (히스토그램, 밀집도)

pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) 
plt.show( )
```

### 4-4. 데이터프레임과 시리즈로 그래프 그리기

```python
# 1. 히스토그램 그리기 : series.plot.hist()
ax = plt.subplots()
ax = tips['total_bill'].plot.hist()
```

```python
# 2. tips 데이터프레임에서 2개의 시리즈를 선택하여 히스토그램 그리기
# alpha : 투명도, bins : x축의 간격
fig, ax = plt.subplots()              # 2개의 시리즈를 활용했으므로 fig, ax 2개를 변수로 지정
ax = tips[['total_bill', 'tip']].plot.hist(alpha=0.5, bins=20, ax=ax)
```

```python
# 3. 밀집도, 산점도 그래프, 육각 그래프 그리기

# 밀집도 : series.plot.kde()
ax = plt.subplots()
ax = tips['tip'].plot.kde()
ax.set_title('Density Function of Tip')

# 산점도 : dataframe.plot.scatter()
fig, ax = plt.subplots()
ax = tips.plot.scatter(x='total_bill', y='tip', ax=ax)
ax.set_title('Total Bills vs. Tip')

# 육각 그래프 : dataframe.plot.hexbin()
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', ax=ax)
ax.set_title('Hexagonal Graph of Total bill vs. Tip')
```

```python
# 4. 육각 그래프의 육각형 크기 조절하기 : dataframe.plot.hexbin(gridsize=10)
# gridsize 파라미터를 활용해 조절
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=ax)
```

### 4-5. **`seaborn`** 라이브러리로 **`그래프 스타일`** 설정하기
  - seaborn 라이브러리 스타일
    : darkgrid, whitegrid, dark, white, ticks

```python
# 1. 기본적인 violinplot 그리기
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# 2. whitegrid로 스타일 설정하기
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# 3. for문을 이용해 그래프에 모든 스타일을 하나씩 적용하기 
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
